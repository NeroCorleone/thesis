```python
import numpy as np
import scipy.ndimage
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
%matplotlib inline
```

```python
cmap = mpl.cm.viridis
norm = mpl.colors.Normalize(vmin=0, vmax=1)

fontlabelsize = 24
ticklabelsize = 20
```

### Full gate covering scattering region except for narrow edge channel

```python
designfile = '/home/nefta/thesis/designfiles/topgate_full_edges.png'
topgate = 1 - scipy.ndimage.imread(designfile, mode='L').T / 255
topgate_gauss = np.fliplr(scipy.ndimage.gaussian_filter(topgate, 0))
scattering_region = np.fliplr(1 - scipy.ndimage.imread(
    '/home/nefta/thesis/designfiles/scattering_region.png', mode='L').T / 255)
plt.imshow(topgate_gauss + scattering_region)
```

```python
edges_dir = '/home/nefta/thesis/results/edges/supercurrent/'
edges_subdirs = sorted(list(os.walk(edges_dir))[0][1])
vsg_vals = [-0.01, -0.02, -0.03, -0.04, -0.045, -0.05, -0.055, -0.06, -0.07, -0.08, -0.09]
```

```python
cdelta = round(cmap.N  / len(vsg_vals))
```

```python
list(enumerate(zip(vsg_vals, edges_subdirs)))
```

```python
fig, ax = plt.subplots(figsize=(16,9))
for index, (vsg, subdir) in list(enumerate(zip(vsg_vals, edges_subdirs)))[2::2]:
    data = edges_dir  +  subdir + '/data.csv'
    ic = pd.read_csv(data, header=None, delimiter=' ')[0].values
    ic_zero = np.amax(ic)
    ax.plot(ic/ic_zero, label=r'$\varphi_{{SG}} = {}$'.format(vsg), linewidth=3)#, color=cmap(index*cdelta))
ax.legend(prop={'size': 18})
```

### QPC-like gate width split at the edges for transsmission

```python
designfile = '/home/nefta/thesis/designfiles/full_gate_both_edges.png'
topgate = scipy.ndimage.imread(designfile, mode='L').T / 255
topgate_gauss = 1 - np.fliplr(scipy.ndimage.gaussian_filter(topgate, 0))
scattering_region = np.fliplr(1 - scipy.ndimage.imread(
    '/home/nefta/thesis/designfiles/scattering_region.png', mode='L').T / 255)
plt.imshow(topgate_gauss + scattering_region)
```

```python
fullgate_edges_dir = '/home/nefta/thesis/results/full_gate_edges/supercurrent/'
fullgate_edges_subdirs = sorted(list(os.walk(fullgate_edges_dir))[0][1])
vsg_vals = [-0.01, -0.02, -0.03, -0.04, -0.04, -0.05, -0.06, 
              -0.07, -0.08, -0.09, -0.1, -0.11, -0.12, -0.13, -0.14]
```

```python
fig, ax = plt.subplots(figsize=(16,9))
for vsg, subdir in list(zip(vsg_vals, fullgate_edges_subdirs))[2::3]:
    data = fullgate_edges_dir  +  subdir + '/data.csv'
    ic = pd.read_csv(data, header=None, delimiter=' ')[0].values
    ic_zero = np.amax(ic)
    ax.plot(ic/ic_zero, label=r'$\varphi_{{SG}} = {}$'.format(vsg), linewidth=3,)
ax.legend(prop={'size': 18})
```
