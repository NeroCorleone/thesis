```python
import matplotlib.pyplot as plt
import numpy as np
import csv
%matplotlib inline
```

```python
def read_imag_values(file):
    result = []
    with open(file) as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        #next(reader)
        for index, row in enumerate(reader):
            #print(row)
            result.append([float(element) for element in row])
    return(result)
```

```python
phi0 = 2.067e-15 #in T*m**2
```

```python
def transform_to_density(B, real_values, imag_values, y_values):
    phi0 = 2.067e-15 #in T*m**2
    L = 100#1e-6
    W = 360#1.8e-6 #in m
    
    #omega = (2 * np.pi * L) 
    omega = (2 * np.pi) /  W
    phi_values = B * W * L
    dPhi = phi_values[1] - phi_values[0]
    
    #dB = B[1] - B[0]
    #real, imag = Ic
    
    Jy_values = []
    for y in y_values:
        Jy_real = 0
        Jy_imag = 0
        for k, phi in enumerate(phi_values):
            Ic_real = real_values[k] #Fourier coefficient
            Ic_imag = imag_values[k]
            Jy_real += (Ic_real * np.cos(omega * phi * y) - Ic_imag * np.sin(omega * phi * y)) * dPhi 
            Jy_imag += (Ic_imag * np.cos(omega * phi * y) + Ic_real * np.sin(omega * phi * y)) * dPhi
        Jy_abs = np.sqrt(Jy_real**2 + Jy_imag**2)
        Jy_values.append(Jy_abs)
    return(Jy_values)
```

```python
maxB = 9e-05
magnetic_field = np.linspace(-maxB, maxB, 1000)
area = 100 * 360
values = np.asarray(read_imag_values(datafile)).T
phi_norm = (area * magnetic_field) /phi0
```

```python
fig, ax = plt.subplots(figsize=(16,9))
for index, labelname in enumerate(['abs', 'real', 'imag']):
    ax.plot(phi_norm, values[index], label=labelname, linestyle='none', marker='o')
ax.legend()
```
